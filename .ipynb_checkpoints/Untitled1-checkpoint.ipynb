```python
import pandas as pd
```

```python
acs_2801_pct_df=pd.read_csv('acs_s2801_USETHIS.csv',low_memory=False)
```

```python
acs_2801_pct_df.head()
```

```python
zip_state_county_df=pd.read_csv('zip_to_state_and_county.csv')
```

```python
zip_state_county_df.head()
```

```python
acs_2801_merge = pd.merge(acs_2801_pct_df,zip_state_county_df, left_on='ZIP',right_on='ZCTA5')
```

```python
acs_2801_merge.head()
```

```python
acs_2801_merge=acs_2801_merge[acs_2801_merge['STATE'] == 47] 
```

```python
acs_2801_merge.head()
```

```python
school_df = pd.read_csv('school_info_combined.csv')
```

```python
acs_2801_merge=pd.merge(acs_2801_merge,school_df,left_on='ZCTA5',right_on='ZIP')
```
